```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
airline=pd.read_csv('C:\\Users\\bharg\\Desktop\\Bhargav\Airline_review.csv')
airline
```

```python
airline.duplicated().sum()
```

```python
airline.isnull().sum()
```

```python
numerical=[fea for fea in airline.columns if airline[fea].dtypes!=object ]
numerical
```

```python
categorical=[fea for fea in airline.columns if airline[fea].dtypes==object]
categorical
```

```python
numerical=airline[numerical]
```

```python
categorical=airline[categorical]
```

```python
numerical
```

```python
categorical
```

```python
airline_filled_numerical_data=numerical.fillna(numerical.mean())
airline_filled_numerical_data
```

```python
airline_filled_categorical_data=categorical.fillna(categorical.mode().iloc[0])
airline_filled_categorical_data
```

```python
airline_filled_numerical_data.isnull().sum()
```

```python
airline_filled_categorical_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
airline_filled_categorical_data['Recommended']=le.fit_transform(airline_filled_categorical_data['Recommended'])
```

```python
airline_filled_categorical_data['Recommended']
```

```python
list(le.classes_)
```

```python
x=airline_filled_numerical_data[['Seat Comfort','Cabin Staff Service']]
x
```

```python
y=airline_filled_categorical_data['Recommended']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
```

```python
from collections import Counter
Counter(y_train)
```

# This is Balanced data

```python
from sklearn.svm import SVC
```

```python
s=SVC(kernel='rbf',C=100,gamma=10)
```

```python
s.fit(x_train,y_train)
```

```python
y_predicted=s.predict(x_test)
y_predicted
```

```python
s.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix=confusion_matrix(y_test,y_predicted)
matrix
```

```python
sns.heatmap(matrix,annot=True,fmt='.1f',xticklabels=['no', 'yes'],yticklabels=['no', 'yes'])
plt.xlabel('Predicted class')
plt.ylabel('True class')
```
